```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_parquet('data/1.parquet')
data
```

```python
plt.figure(figsize = (15,15))
plt.plot(data.rtctime, data.target_temperature)
plt.plot(data.rtctime, data.feature_c)
plt.plot(data.rtctime, data.feature_ct)
plt.plot(data.rtctime, data.ambient_temp)
plt.plot(data.rtctime, data.car_speed)
plt.plot(data.rtctime, data.soc)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (15,15))
plt.plot(data.rtctime, data.feature_motorspeed)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.plot(data.soc, data.ambient_temp)
plt.plot(data.soc, data.target_temperature)
plt.plot(data.soc, data.feature_c)
plt.plot(data.soc, data.feature_ct)
plt.legend()
plt.show()
```

```python
corr_data = data[['target_temperature', 'feature_c', 'feature_ct', 'feature_motorspeed', 'ambient_temp', 'car_speed', 'soc']].corr()
sns.heatmap(corr_data, annot=True)
```
