```python
import os
import pandas as pd
import pyarrow.parquet as pq

# Path to the folder containing the Parquet files
folder_path = 'data/'

# Get a list of all Parquet files in the folder
file_list = [file for file in os.listdir(folder_path) if file.endswith('.parquet')]

# Initialize an empty list to store individual DataFrames
dfs = []

# Read and append each Parquet file to the list
for file in file_list:
    file_path = os.path.join(folder_path, file)
    df = pd.read_parquet(file_path, engine='pyarrow')
    print(df.shape)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
full_df = pd.concat(dfs)

# Print the resulting DataFrame
full_df
```
